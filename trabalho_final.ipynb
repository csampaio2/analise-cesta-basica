```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('cesta_basica_1994_2025.xls')
df.head()
```

```python
df.describe()
```

## Filtrando as capitais que não possuem valores de cesta basica faltante

```python

# Contar valores não nulos por coluna
non_null_counts = df.notnull().sum()

# Ordenar as colunas por quantidade de valores preenchidos
sorted_columns = non_null_counts.sort_values(ascending=False)
print(sorted_columns)

# Filtrandro apenas as capitais com mais dados
df_373 = df.loc[:, df.notnull().sum() >= 373]
df_373.head()
```

```python
# Removendo linhas vazias
df_top_data = df_373[:-3]

# Apagando caracteres em branco da coluna
df_top_data.columns = df_top_data.columns.str.strip()

# Removendo linha de periodo duplicado
df_top_data = df_top_data.drop(258)

# Transformando o periodo para o formato datetime
df_top_data['Periodo'] = pd.to_datetime(df_top_data['Periodo'], format='%m-%Y')

# Definindo novo indice do dataframe
df_top_data = df_top_data.set_index('Periodo')

# Agrupando os dados no periodo de 12 meses
df_top_data_12M = df_top_data.resample('12ME').mean()
```

## Gráfico de linhas da evolução dos preços das cestas básicas de 12 em 12 meses de 1994 a 2025

```python
df_top_data_12M.plot(figsize=(15,7))

plt.title("Evolução do preço das cestas básicas de 1994 a 2025")
plt.ylabel('Valor da cesta básica em R$')
plt.xlabel('Período (12 em 12 meses)')
plt.grid(True)
plt.show()
```

## Mapa de calor dos preços das cestas básica das capitais

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_top_data_12M.T, cmap="YlOrRd", annot=False, cbar_kws={'label': 'Valor da Cesta Básica (R$)'})
plt.title("Mapa de Calor do Custo da Cesta Básica nas Capitais Brasileiras (1994-2025)")
plt.xlabel("Ano")
plt.ylabel("Cidades")
plt.tight_layout()


plt.show()
```

## Boxplot
### Compara a dispersão, mediana e possíveis outliers dos valores entre as cidades.

```python

plt.figure(figsize=(14, 8))
sns.boxplot(data=df_top_data_12M, orient='h')
plt.title("Distribuição dos Valores da Cesta Básica por Capital (1994-2025)")
plt.xlabel("Valor da Cesta Básica (R$)")
plt.ylabel("Cidade")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
